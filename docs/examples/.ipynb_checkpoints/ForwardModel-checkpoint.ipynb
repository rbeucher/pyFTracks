```python
import pyTracks
from pyTracks.thermal_history import WOLF1
```

```python
model = pyTracks.KetchamModel(WOLF1, 
                              initial_track_length=16.3,
                              kinetic_parameter_type="ETCH_PIT_LENGTH",
                              kinetic_parameter_value=1.65
                             )
```

```python
model.get_initial_track_length()
```

```python
model.solve()
```
